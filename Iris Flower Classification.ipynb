# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# Loading Dataset

```python
columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Species"]

df = pd.read_csv("iris.data.csv", names=(columns))
```

```python
df.head()
```

# Visualization

### Pairplot

```python
#Plotting Pairplot for Visualizing Relationships between features.

sns.pairplot(df, hue="Species")
```

### Boxplot

```python
# Plotting Boxplots to find OUTLIERS.

sns.boxplot(data=df)
```

# Extracting Input and Output Columns

```python
# Scaling the Features.

# In the dataset,
# Inputs(Features) = Sepal Length, Sepal Width, Petal Length, Petal Width.
# Output(Target) = Species

# x = Inputs = Sepal Length, Sepal Width, Petal Length, Petal Width.
# y = Output = Species
```

```python
x = df[["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]].values        # Features
y = df["Species"].values                                                             # Target
```

```python
x
```

```python
y
```

# Split the dataset into Train and Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# Initialize KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=3)
```

# Train the Model

```python
knn_classifier.fit(x_train, y_train)
```

# Make Predictions on the Test Set

```python
y_pred = knn_classifier.predict(x_test)
```

```python
y_pred
```

# Confusion Matrix for Visualization

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=df['Species'].unique(), yticklabels=df['Species'].unique())
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# Evaluate the Model

```python
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
```

# Results

```python
print(f"Accuracy: {accuracy:.2f}")
```

```python
print("Classification Report:\n\n",classification_report(y_test, y_pred))
```
